```python
import sys
import os
sys.path.append(os.path.abspath('C:/Users/vpming/tuni_ml/src'))
from extract_data import build_features_10samples_df, extract_noise
```

```python
data_path = 'C:/Users/vpming/tuni_ml/data'
df = build_features_10samples_df(data_path)
df['noise'] = df['simulation_file'].apply(extract_noise)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['dis_to_target'], df['cVariance'], alpha=0.5, s=12)
plt.xlabel('Distance to Target')
plt.ylabel('cVariance')
plt.title('cVariance vs Distance to Target')

plt.subplot(1, 2, 2)
plt.scatter(df['dis_to_target'], df['cMax'], alpha=0.5, s=12, color='orange')
plt.xlabel('Distance to Target')
plt.ylabel('cMax')
plt.title('cMax vs Distance to Target')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# cVariance vs Distance to Target, colored by noise
plt.subplot(1, 3, 1)
sns.scatterplot(df, x='dis_to_target', y='cVariance', hue='noise', palette='viridis', s=15, alpha=0.7)
plt.xlabel('Distance to Target')
plt.ylabel('cVariance')
plt.title('cVariance vs Distance, by Noise')
plt.legend(title="Noise", bbox_to_anchor=(1,1))

# cMax vs Distance to Target, colored by noise
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='dis_to_target', y='cMax', hue='noise', palette='viridis', s=15, alpha=0.7)
plt.xlabel('Distance to Target')
plt.ylabel('cMax')
plt.title('cMax vs Distance, by Noise')
plt.legend(title="Noise", bbox_to_anchor=(1,1))

plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='cVariance', y='cMax', hue='noise', palette='viridis', s=15, alpha=0.7)
plt.xlabel('cVariance')
plt.ylabel('cMax')
plt.title('cVariance vs cMax, by Noise')
plt.legend(title="Noise", bbox_to_anchor=(1,1))

plt.tight_layout()
plt.show()
```
